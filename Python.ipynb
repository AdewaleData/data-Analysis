```python
# Reading the CSV file into a Pandas DataFrame
df = pd.read_csv('clean.csv')

# Printing the first few rows of the DataFrame
print(df.head())
```

```python
# VISUALIZING THE FREQUENCY OF THE POTENTIAL ENERGY RATING FOR 
# DIFFRENT PROPERTIES TO SEE THE RATING THAT MOST PROPERTIES 
# HAVE IN THE DATASET

# the order of the ratings

A='very efficient- lower running costs'
B=''
C=''
D='not too bad'
E=''
F=''
G="not energy efficient- higher running cost"

# Group the data by property type and rating, and count the frequency
grouped_data = df.groupby(['PROPERTY_TYPE', 'POTENTIAL_ENERGY_RATING'])['POTENTIAL_ENERGY_RATING'].count()

# Reshape the data into a pivot table
pivot_table = grouped_data.unstack()

# Plot a grouped bar chart for each property type
pivot_table.plot(kind='bar')

# Add labels and titles
plt.xlabel('Property Type')
plt.ylabel('Frequency')
plt.title('Frequency of potential energy Ratings by Property Type')
plt.legend(title='potential energy Rating')

# Show the plot
plt.show()

# FROM THIS PLOT BELOW, 
#the
```

```python
# VISUALIZING THE FREQUENCY OF THE CURRENT ENERGY RATING FOR 
# DIFFRENET PROPERTIES TO SEE THE RATING THAT MOST PROPERTIES 
# HAVE IN THE DATASET


# Group the data by property type and rating, and count the frequency
grouped_data = df.groupby(['PROPERTY_TYPE', 'CURRENT_ENERGY_RATING'])['CURRENT_ENERGY_RATING'].count()

# Reshape the data into a pivot table
pivot_table = grouped_data.unstack()

# Plot a grouped bar chart for each property type
pivot_table.plot(kind='bar')

# Add labels and titles
plt.xlabel('Property Type')
plt.ylabel('Frequency')
plt.title('Frequency of current energy Ratings by Property Type')
plt.legend(title='Current energy Rating')

# Show the plot
plt.show()
```

```python
# Group the data by the column with groups, and count the frequency
grouped_data = df.groupby('ROOF_ENERGY_EFF')['ROOF_ENERGY_EFF'].count()

# Plot a pie chart
plt.pie(grouped_data.values, labels=grouped_data.index, autopct='%1.1f%%')

# Add title
plt.title('Pie Chart of the roof energy efficiency ratings ')

# Show the plot
plt.show()
```

```python
# Group the data by the column with groups, and count the frequency
grouped_data = df.groupby('LIGHTING_ENERGY_EFF')['LIGHTING_ENERGY_EFF'].count()

# Plot a pie chart
plt.pie(grouped_data.values, labels=grouped_data.index, autopct='%1.1f%%')

# Add title
plt.title('Pie Chart of the main lighting energy efficiency ratings ')

# Show the plot
plt.show()
```
